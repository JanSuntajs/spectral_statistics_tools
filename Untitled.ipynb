```python
#  prepare

import os
import sys
import glob
import numpy as np
from utils import unfolding as unf
from utils import misc as misc
from sff import sff_functions as sff
from scipy.special import comb
from matplotlib import pyplot as plt 
load_path = './example_data/L_14_Nu_7/'

def load_data(path):
    
    cwd = os.getcwd()
    
    os.chdir(path)
    
    files = glob.glob('Eigvals*')
    files = np.array([np.fromfile(file) for file in files])
    os.chdir(cwd)
    return files
```

```python
# load data

# ergodic case
spec_erg = load_data(load_path + 'ergodic/')
nener = spec_erg.shape[1]

spectral_width = (0., 1.)
# unfold data
spec_unfold = unf.unfold_function(spec_erg, n=3, spectral_width=spectral_width)
spec_unfold = unf.correct_slope(spec_unfold)

# get spectral data
mean_ener, sq_ham_tr, ham_tr_sq, gamma = misc.get_ham_misc(spec_unfold, spectral_width=spectral_width)

# filtering data
filter_, dims, dims_eff, normal_con, normal_uncon = misc.spectral_filtering(spec_unfold, misc.gaussian_filter, 0.5, np.sqrt(gamma), mean_ener)
# filter_ = np.ones_like(spec_unfold, dtype=np.float)
# dims_eff = spec_unfold.shape[1]
# normal_con = 1.
# normal_uncon = 1.
# obtain sff
taulist = np.logspace(-4,1.5,500)
sfflist = sff.sff_spectra(spec_unfold, taulist, filter_)
```

```python
sff_full = np.mean(np.abs(sfflist)**2, axis=0)
sff_uncon = np.abs(np.mean(sfflist, axis=0))**2
```

```python
plt.loglog(taulist/ (2*np.pi), (sff_full - (normal_con/normal_uncon)*sff_uncon) / dims_eff); 
# plt.ylim(0.1,100)
plt.axhline(y=1)
plt.show()
```

```python
plt.loglog(taulist/ (2*np.pi), (sff_full) / dims_eff); 
# plt.ylim(0.1,100)
plt.axhline(y=1)
plt.show()
```
